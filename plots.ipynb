```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# Carregar os dados do arquivo CSV
df = pd.read_csv('results_12_sitios_VQC.csv')

# Extrair os dados das colunas
g = df.iloc[:, 0]
kappa = df.iloc[:, 1]
fase = df.iloc[:, 2]

# Melhorar o esquema de cores para visualizao
fase_colors = {0: '#1f77b4',  # Azul escuro
               1: '#ff7f0e',  # Laranja
               2: '#2ca02c'}  # Verde
colors = [fase_colors[f] for f in fase]

# Funes Ising e BKT
def g_I(kappa):
    return (1 - kappa) / kappa * (1 - np.sqrt((1 - 3 * kappa + 4 * kappa**2) / (1 - kappa)))

def g_BKT(kappa):
    return 1.05 * np.sqrt((kappa - 0.5) * (kappa - 0.1))

# Criar o grfico
plt.figure(figsize=(12, 8))

# Plotar os pontos com cores diferentes para cada fase, sem borda, pontos maiores e sem transparncia
scatter = plt.scatter(kappa, g, c=colors, edgecolor='none', s=100)

# Adicionar as funes Ising e BKT
kappa_range_is = np.linspace(0, 0.5, 500)
kappa_range_bkt = np.linspace(0.5, 1, 500)

plt.plot(kappa_range_is, g_I(kappa_range_is), 'r-', label=r'$g_{I}(\kappa)$', linewidth=2)
plt.plot(kappa_range_bkt, g_BKT(kappa_range_bkt), 'b--', label=r'$g_{BKT}(\kappa)$', linewidth=2)

# Ajustar e formatar o grfico
plt.xlabel(r'$\kappa$', fontsize=30)  # Aumentado o tamanho dos labels dos eixos
plt.ylabel(r'$g$', fontsize=30)        # Aumentado o tamanho dos labels dos eixos
plt.xticks(fontsize=22)                # Aumentado o tamanho dos nmeros dos eixos
plt.yticks(fontsize=22)                # Aumentado o tamanho dos nmeros dos eixos

# Adicionar a legenda com identificao das cores e sem transparncia
legend_elements = [mlines.Line2D([0], [0], marker='o', color='w', markerfacecolor=fase_colors[0], markersize=15, label='ferro'),
                   mlines.Line2D([0], [0], marker='o', color='w', markerfacecolor=fase_colors[1], markersize=15, label='para'),
                   mlines.Line2D([0], [0], marker='o', color='w', markerfacecolor=fase_colors[2], markersize=15, label='anti')]

plt.legend(handles=legend_elements + [plt.Line2D([0], [0], color='r', lw=2, label=r'$g_{I}(\kappa)$'),
                                       plt.Line2D([0], [0], color='b', lw=2, linestyle='--', label=r'$g_{BKT}(\kappa)$')],
           title='VQC - N=12', title_fontsize='18', fontsize='14', frameon=True, framealpha=1)  # Sem transparncia na legenda

# Remover a grade
plt.grid(False)

# Salvar o grfico como um arquivo PDF
plt.tight_layout()
plt.savefig('grafico_resultado.pdf', format='pdf')

# Mostrar o grfico
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# Funes Ising e BKT
def g_I(kappa):
    return (1 - kappa) / kappa * (1 - np.sqrt((1 - 3 * kappa + 4 * kappa**2) / (1 - kappa)))

def g_BKT(kappa):
    return 1.05 * np.sqrt((kappa - 0.5) * (kappa - 0.1))

# Novo esquema de cores para visualizao
fase_colors = {0: '#1f77b4',  # Azul claro
               1: '#ff7f0e',  # Laranja
               2: '#2ca02c'}  # Verde

# Lista de arquivos e ttulos
files = [
    ('Salva_Certo_PF_8_sites_VQC.csv', 'VQC - N=8', True),  # Adicionado 'True' para indicar inverso
    ('results_12_sitios_VQC.csv', 'VQC - N=12', False),
    ('QSVM_8_sites_pred.csv', 'QSVM - N=8', True),  # Adicionado 'True' para indicar inverso
    ('results_12_sitios_QSVM.csv', 'QSVM - N=12', False)
]

# Criar subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12), sharex=True, sharey=True)

for ax, (file, title, invert_axes) in zip(axes.flatten(), files):
    # Carregar os dados do arquivo CSV
    df = pd.read_csv(file)

    # Corrigir rotulagem das fases para arquivos N=8
    if 'QSVM_8_sites_pred' in file:
        df.iloc[:, 2] = df.iloc[:, 2] - 1

    # Extrair os dados das colunas
    g = df.iloc[:, 0]
    kappa = df.iloc[:, 1]
    fase = df.iloc[:, 2]

    # Mapear as fases para cores
    colors = [fase_colors[f] for f in fase]

    if invert_axes:
        # Inverter eixos para os casos N=8
        scatter = ax.scatter(g, kappa, c=colors, edgecolor='none', s=100)
        ax.set_xlabel(r'$\kappa$', fontsize=28)
        ax.set_ylabel(r'$g$', fontsize=28)
        ax.plot(np.linspace(0, 0.5, 500), g_I(np.linspace(0, 0.5, 500)), 'r-', label=r'$g_{I}(\kappa)$', linewidth=2)
        ax.plot(np.linspace(0.5, 1, 500), g_BKT(np.linspace(0.5, 1, 500)), 'b--', label=r'$g_{BKT}(\kappa)$', linewidth=2)
    else:
        scatter = ax.scatter(kappa, g, c=colors, edgecolor='none', s=100)
        ax.set_xlabel(r'$\kappa$', fontsize=28)
        ax.set_ylabel(r'$g$', fontsize=28)
        ax.plot(np.linspace(0, 0.5, 500), g_I(np.linspace(0, 0.5, 500)), 'r-', label=r'$g_{I}(\kappa)$', linewidth=2)
        ax.plot(np.linspace(0.5, 1, 500), g_BKT(np.linspace(0.5, 1, 500)), 'b--', label=r'$g_{BKT}(\kappa)$', linewidth=2)

    # Adicionar textos s fases (posies fixas em todos os grficos)
    ax.text(0.20, 0.25, 'ferro', fontsize=20, color='black', ha='center', transform=ax.transAxes)
    ax.text(0.5, 0.5, 'para', fontsize=20, color='black', ha='center', transform=ax.transAxes)
    ax.text(0.8, 0.25, 'anti', fontsize=20, color='black', ha='center', transform=ax.transAxes)

    # Ajustar e formatar o grfico
    ax.set_title(title, fontsize=28)
    ax.tick_params(axis='both', which='major', labelsize=22)
    ax.grid(False)

# Adicionar a legenda com identificao das cores e sem transparncia
legend_elements = [mlines.Line2D([0], [0], marker='o', color='w', markerfacecolor=fase_colors[0], markersize=15, label='ferro'),
                   mlines.Line2D([0], [0], marker='o', color='w', markerfacecolor=fase_colors[1], markersize=15, label='para'),
                   mlines.Line2D([0], [0], marker='o', color='w', markerfacecolor=fase_colors[2], markersize=15, label='anti')]

# Adicionar a legenda fora dos subplots
fig.legend(handles=legend_elements + [plt.Line2D([0], [0], color='r', lw=2, label=r'$g_{I}(\kappa)$'),
                                       plt.Line2D([0], [0], color='b', lw=2, linestyle='--', label=r'$g_{BKT}(\kappa)$')],
           loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=5, fontsize=28, frameon=True, framealpha=1, markerscale=1.5)

# Ajustar layout e salvar
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Ajustar o rect para dar espao para a legenda
plt.savefig('subplots_resultado_corrigido_com_texto.pdf', format='pdf', bbox_inches='tight')

# Mostrar o grfico
plt.show()
```
